```python
import numpy as np
import matplotlib.pyplot as plt

Te=0.005
t=np.arange(-4,4+Te,Te)
N=len(t)

tab_color = ['m','r','g']

for i in range (3):
    sigma = [ 1/2**k for k in range (2,8,2)]
    g = (np.exp(-t**2 / (2 * sigma[i]**2))/ (sigma[i] * np.sqrt(2 * np.pi)) )
    plt.subplot(2,3,i+1)
    plt.plot(t,g,tab_color[i])
    nu_e = 1/Te 
    nu = np.linspace(-nu_e/2,nu_e/2,N)
    tf = np.zeros(N,dtype=np.complex128)
    for k in range (N):
        tf[k]=np.trapz(g*np.exp(-2*1j*np.pi*nu[k]*t),t)
    plt.subplot(2,3,i+4)
    plt.plot(nu,tf.real,tab_color[i])

```
